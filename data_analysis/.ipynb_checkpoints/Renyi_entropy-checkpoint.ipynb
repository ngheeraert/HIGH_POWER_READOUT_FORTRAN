## Mean number of photons versus time

```python
#-- plotting the mean photon number in the cavity
for ad in sorted_keys:
    plt.plot( dyns[ad].times, dyns[ad].nk[:,0], label='ad='+str(np.round(ad*0.001,3))  )
plt.ylabel( "$<n>$" )
plt.xlabel( "$t$" )
plt.title( "photons in cavity VS time"  )
plt.legend(loc='upper left')
plt.show()
```

## Renyi entropy between qubit and cavity

```python
entropy = {}
nk_dict = {}
for ad in dyns:
    nk_dict[ad] = dyns[ad].nk[-1,0]

for ad in dyns:
    if ad>15:
        entropy[ ad ] = dyns[ad].final_state.renyi_entropy(0, 30)
    if ad<15:
        entropy[ ad ] = dyns[ad].final_state.renyi_entropy(0, 15)
```

```python
for ad in entropy_vs_time_keys:
    plt.plot( dyns[ad].times, dyns[ad].entropy, label='ad='+str(np.round(ad*0.001,3))  )
plt.ylabel( "$S_{q \oplus cav}$" )
plt.xlabel( "$t$" )
plt.title( "Renyi entropy VS time"  )
plt.legend()
```

```python
for ad in sorted_keys:
    plt.scatter( nk_dict[ad], entropy[ad],label='ad='+str(np.round(ad*0.001,3)) )

plt.xlim( -1,11 )
plt.title( 'Entropy versus mean number of photons in the cavity \n at the end of the simulation.' )
plt.ylabel( "$S_{q \oplus cav}$" )
plt.xlabel( r'$\langle a^\dag_{cav}a_{cav}\rangle$ -- Mean number photon in the cavity at t=80' )
plt.show()
```
