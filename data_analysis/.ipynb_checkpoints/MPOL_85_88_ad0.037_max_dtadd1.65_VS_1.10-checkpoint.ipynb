```python
import dynamics_analysis as da
import matplotlib.pyplot as plt
from copy import deepcopy
import numpy as np

%load_ext autoreload
%autoreload 2
```

```python
path = '../../data_fortran/MPOL_DCT_85/data/'
path88 = '../../data_fortran/MPOL_DCT_88/data/'

paramchar_n44 = "nl2_wc5.500_n2_50_E0.10E-09_dtadd1.65_ar0.0_wq5.000_g0.050_gam0.016000_nm250_mr0.00_p0.5E-05_bw1.500_bc5.250_wd5.500_ad0.037_anh159.155_dt0.0100_errl0.1E-03_qb1_cv0.0_tmax80"
dyn_n44 = da.dynamics(2,path, paramchar_n44)

paramchar_n40_1p1 = "nl2_wc5.500_n2_40_E0.10E-09_dtadd1.10_1.79_12_ar0.0_wq5.000_g0.050_gam0.016000_nm250_mr0.00_p0.5E-05_bw1.500_bc5.250_wd5.500_ad0.037_anh159.155_dt0.0100_errl0.1E-03_qb1_cv0.0_tmax80"
dyn_n40_1p1 = da.dynamics(2,path88, paramchar_n40_1p1)

paramchar_n2 = "nl2_wc5.500_n2_2_E0.10E-08_dtadd0.01_ar0.0_wq5.000_g0.050_gam0.016000_nm250_mr0.00_p0.5E-05_bw1.500_bc5.250_wd5.500_ad0.037_anh159.155_dt0.0120_errl0.1E-03_qb1_cv0.0_tmax80"
dyn_n2 = da.dynamics(2,path, paramchar_n2)
```

## Mean number of photons versus time

```python
plt.plot( dyn_n44.times, dyn_n44.nk[:,0], label='ncs=44'  )
plt.plot( dyn_n40_1p1.times, dyn_n40_1p1.nk[:,0], label='ncs=40_1p1'  )
plt.plot( dyn_n2.times, dyn_n2.nk[:,0], label='ncs=2'  )
plt.ylabel( "$<n>$" )
plt.xlabel( "t" )
plt.title( "photons in cavity VS time"  )
plt.legend()
plt.show()

plt.plot( dyn_n44.times, dyn_n44.error, label='ncs=44'  )
plt.plot( dyn_n40_1p1.times, dyn_n40_1p1.error, label='ncs=40_1p1'  )
plt.plot( dyn_n2.times, dyn_n2.error, label='ncs=2'  )
plt.ylabel( "$Error$" )
plt.xlabel( "t" )
plt.title( "Error VS time"  )
plt.yscale( 'log' )
plt.legend()
plt.show()
```

## **Bath photon spectrum**

```python
plt.plot( dyn_n2.wk/(2*(np.pi)), dyn_n2.nk[-1,1:], label='ncs=2' , c='gray',alpha=0.5 )
plt.plot( dyn_n44.wk/(2*(np.pi)), dyn_n44.nk[-1,1:], label='ncs=44'  )
plt.plot( dyn_n40_1p1.wk/(2*(np.pi)), dyn_n40_1p1.nk[-1,1:], label='ncs=40_1p1'  )

plt.title( "Full spectrum (mean number of photons)"  )
plt.xlabel( "Freqs $\omega/(2\pi)$"  )
plt.ylabel( "$<n_k>$" )
plt.legend()

plt.xlim(4.8,5.7)
plt.ylim( 0,0.005 )
```

```python
one_photon_nk = dyn_n44.final_state.one_photon_k_nb()
two_photon_nk = dyn_n44.final_state.two_photon_k_nb()
three_photon_nk = dyn_n44.final_state.three_photon_k_nb()
```

```python
#plt.plot( dyn_n40_1p1.wk/(2*(np.pi)), dyn_n40_1p1.nk[-1,1:], label='<n_k>'  )
plt.plot( dyn_n44.wk/(2*(np.pi)), one_photon_nk[1:], label='|1>'  )
plt.plot( dyn_n44.wk/(2*(np.pi)), two_photon_nk[1:], label='|2>'  )
plt.plot( dyn_n44.wk/(2*(np.pi)), three_photon_nk[1:], label='|3>'  )
plt.title( "spectrum decompostion"  )
plt.xlabel( "Freqs $\omega/(2\pi)$"  )
plt.ylabel( "$<n_k>$" )
plt.legend()
plt.xlim(4.8,5.7)
plt.ylim( 0,1e-10 )
```

# **Applying destruction operator at 4.95 GHz**

```python
target_mode=75
print("-- Frequency of mode number "+str(target_mode)+ ' =',dyn_n40_1p1.wk[ target_mode ]/(2*np.pi))
```

```python
a_st_n40_1p1 = deepcopy( dyn_n40_1p1.final_state )
a_st_n40_1p1.destroy_k(75)
tot_photon_nk_n40_1p1 = dyn_n40_1p1.final_state.tot_photon_k_nb()
a_tot_photon_nk_n40_1p1 = a_st_n40_1p1.tot_photon_k_nb()
```

```python
plt.plot( dyn_n40_1p1.wk/(2*(np.pi)), tot_photon_nk_n40_1p1[1:], label='before destruction'  )
plt.plot( dyn_n40_1p1.wk/(2*(np.pi)), a_tot_photon_nk_n40_1p1[1:], label='after destruction'  )
plt.title( "Qubit emssion spectrum -- before/after destruciton at 4.95 GHz"  )
plt.xlabel( "Freqs $\omega/(2\pi)$"  )
plt.ylabel( "$<n_k>$" )
plt.xlim(4.8,5.7)
plt.ylim( 0,0.005 )
plt.legend()
plt.show()

plt.plot( dyn_n40_1p1.wk/(2*(np.pi)), tot_photon_nk_n40_1p1[1:], label='before destruction'  )
plt.plot( dyn_n40_1p1.wk/(2*(np.pi)), a_tot_photon_nk_n40_1p1[1:], label='after destruction'  )
plt.title( "Cavity emission spectrum -- before/after destruciton at 4.95 GHz"  )
plt.xlabel( "Freqs $\omega/(2\pi)$"  )
plt.ylabel( "$<n_k>$" )
plt.xlim(5.4,5.6)
plt.legend()
plt.show()
```
