```python
wc_dict={}
Gamma_dict={}
Gamma_dict2={}

for bw in dyns_nm1000:

    end_t_ind = len( dyns_nm2000[bw].times[1:] )//2
    opt1, _ = curve_fit( exp_decay, dyns_nm2000[bw].times[:end_t_ind], dyns_nm2000[bw].nk[:end_t_ind,0] )
    Gamma_dict[bw]=opt1[0]

    def lorentzien( w, wc_in, gam_in, A=0.001 ):
        return A*( (gam_in/2)/( (w-wc_in)**2 + ((gam_in/2)*(w/wc_in)**2) ) )**2

    fig, (ax1,ax2) = plt.subplots( 1,2, figsize=[8,3.5] )

    ax1.plot( dyns_nm2000[bw].times[1:end_t_ind], dyns_nm2000[bw].nk[1:end_t_ind,0], linewidth=10, alpha=0.8,label='data nm=2000'  )
    ax1.plot( dyns_nm1000[bw].times[1:end_t_ind], dyns_nm1000[bw].nk[1:end_t_ind,0], linewidth=6, alpha=0.8,label='data nm=1000'  )
    ax1.plot( dyns_nm2000[bw].times[1:end_t_ind], exp_decay( dyns_nm2000[bw].times[1:end_t_ind], opt1[0]), linewidth=1, c='red', label='fit'  )
    ax1.set_ylabel( "$<n>$" )
    ax1.set_xlabel('t [s]')
    ax1.legend()
    ax1.set_title('Bandwidth='+str(bw)+'GHz \n Decay: $\Gamma/(2\pi)=$ '+str(np.round(1e3*Gamma_dict[bw]/(2*np.pi),5))+' MHz')

    p0=[7.5*2*np.pi,0.95*opt1[0],0.1]
    opt2, _ = curve_fit( lorentzien, dyns_nm2000[bw].wk[:], dyns_nm2000[bw].nk[-1,1:],p0=p0 )
    wc_dict[bw]=opt2[0]
    Gamma_dict2[bw]=opt2[1]

    ax2.scatter( dyns_nm2000[bw].wk[:]/(2*np.pi), dyns_nm2000[bw].nk[-1,1:], label='nm=2000'  )
    ax2.scatter( dyns_nm1000[bw].wk[:]/(2*np.pi), dyns_nm1000[bw].nk[-1,1:]/2, label='nm=1000',s=8  ) 
    ax2.plot( dyns_nm2000[bw].wk[:]/(2*np.pi), \
             lorentzien( dyns_nm2000[bw].wk[:], opt2[0], opt2[1], opt2[2] ), linewidth=1, c='red', label='fit'  )
    plt.axvline( wc_dict[bw]/(2*np.pi) ,c='gray', dashes=[2,2,2,2] )
    ax2.set_xlabel( "Freqs $\omega/(2\pi)$"  )
    ax2.set_ylabel( "$<n_k>$" )
    ax2.set_title( '$\Gamma/(2\pi)=$ '+str(np.round(1e3*Gamma_dict2[bw]/(2*np.pi),5))+' MHz' )
    ax2.set_xlim(7.4,7.6)
    plt.legend()

    plt.tight_layout()
    plt.show()
```
