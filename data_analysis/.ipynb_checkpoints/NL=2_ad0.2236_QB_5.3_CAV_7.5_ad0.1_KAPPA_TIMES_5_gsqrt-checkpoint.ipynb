## Decay rate calibration


```python
path_mac = '../../data_mac/data_wq5304_28_06_2024_gsqrt/'
paramchar_cav1= "nl2_n1_1_E0.10E-01_dtadd0.20_0.20_4_ar0.0_g0.000_gam0.007407_nm250_p0.5E-05_wq5.3037_wc7.500_bc2.300_5.200_6.500_12.600_bw4.000_mr0.00_0.00_0.00_wd7.500_ad0.0000_anh159.155_dt0.0050_errl0.1E-03_lsfe14_qb2_cv1.0_tmax20_TRSM3"
dyn_cav1 = da.dynamics(2, path_mac, paramchar_cav1)
paramchar_cav1_LONG= "nl2_n1_1_E0.10E-01_dtadd0.20_0.20_4_ar0.0_g0.000_gam0.007407_nm250_p0.5E-05_wq5.3037_wc7.500_bc2.300_5.200_6.500_12.600_bw4.000_mr0.00_0.00_0.00_wd7.500_ad0.0000_anh159.155_dt0.0050_errl0.1E-03_lsfe14_qb2_cv1.0_tmax100_TRSM3"
dyn_cav1_LONG = da.dynamics(2,path_mac, paramchar_cav1_LONG)

dyns={}
paramchar= "nl2_n1_1_E0.10E-01_dtadd0.20_0.20_4_ar0.0_g0.250_gam0.007407_nm250_p0.5E-05_wq5.3037_wc7.500_bc2.300_5.200_6.500_12.600_bw4.000_mr0.00_0.00_0.00_wd7.500_ad0.2236_anh159.155_dt0.0050_errl0.1E-03_lsfe14_qb2_cv0.0_tmax30_TRSM3"
dyns[1] = da.dynamics(2,path_mac, paramchar)
paramchar= "nl2_n2_2_E0.10E-01_dtadd0.20_0.20_4_ar0.0_g0.250_gam0.007407_nm250_p0.5E-05_wq5.3037_wc7.500_bc2.300_5.200_6.500_12.600_bw4.000_mr0.00_0.00_0.00_wd7.500_ad0.2236_anh159.155_dt0.0050_errl0.1E-03_lsfe14_qb2_cv0.0_tmax30_TRSM3"
dyns[2] = da.dynamics(2,path_mac, paramchar)
```

```python
from scipy.optimize import curve_fit

def exp_decay(t,gamma):
    return np.exp( -gamma*t )

opt, _ = curve_fit( exp_decay, dyn_cav1.times, dyn_cav1.nk[:,0] )
Gamma=opt[0]

print( '-- fitted decay rate/(2*np.pi) =', Gamma/(2*np.pi), "GHz" )

plt.plot( dyn_cav1.times[1:], dyn_cav1.nk[1:,0], linewidth=8, alpha=0.8,label='data'  )
plt.plot( dyn_cav1_LONG.times[1:], dyn_cav1_LONG.nk[1:,0], linewidth=8, dashes=[2,2,2,2],alpha=0.8,label='data'  )
plt.plot( dyn_cav1.times[1:], exp_decay( dyn_cav1.times[1:], opt[0]), c='red', label='fit'  )
plt.ylabel( "$<n>$" )
plt.xlabel('t [s]')
plt.title('Decay of a coherent state in the cavity with 1 photon on average\n Fitted decay: $\Gamma/(2\pi)=$ '+str(np.round(1e3*Gamma/(2*np.pi),5))+' MHz')
plt.legend()
plt.show()


plt.scatter( dyn_cav1_LONG.wk[:]/(2*np.pi), dyn_cav1_LONG.nk[-1,1:], label='cavity spectrum'  )
plt.xlabel( "Freqs $\omega/(2\pi)$"  )
plt.ylabel( "$<n_k>$" )
plt.legend()
plt.tight_layout()
```
