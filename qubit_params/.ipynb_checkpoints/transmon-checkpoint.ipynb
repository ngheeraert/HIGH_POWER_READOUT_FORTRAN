# Transmon Parameters

We consider the Hamiltonian 
$$
\hat H (t) = 4 E_C \hat n_t^2 - E_J \cos \hat\phi_t +\omega_R \hat a^\dagger \hat a - g \hat n_t (\hat a + \hat a^\dagger) - \epsilon_d (\hat a +\hat a^\dagger) \sin(\omega_d t)
$$

We select the parameters of [1]:

$$
E_J/E_C = 50, \quad E_C/h = 280 \text{ MHz}, \quad \kappa/2\pi = 20 \text{ MHz}, \quad \omega_d = \omega_r
$$



We then reduce to the Rabi Hamiltonian by diagonalization of the transmon and evaluating the charge operator in the its eigenbasis: $ \sum_{i,j \in \{e,g\} } \langle i|\hat n_t |j \rangle $, which is nonzero only on the off-diagonal elements. With this the Hamiltonian becomes

$$
\hat H (t) = \frac{\omega_{eg}}{2} \hat \sigma_z +\omega_R \hat a^\dagger \hat a - g |\langle e|\hat n_t |g \rangle| \sigma_x (\hat a + \hat a^\dagger) - \epsilon_d (\hat a +\hat a^\dagger) \sin(\omega_d t)
$$


1. Shillito, R. et al. Dynamics of Transmon Ionization. Phys. Rev. Applied 18, 034031 (2022).


```python
scq.set_units('GHz')
scq.get_units()
```

```python
Ec = 0.28
Ej = 50*Ec
g = 0.25
```

```python
transmon = scq.Transmon(EJ = Ej,
                        EC = Ec,
                        ng = 0,
                        ncut = 150,
                        truncated_dim = 5) 

mat = transmon.n_operator()
out = mat.view()
print(out)
```

```python
eigenval = transmon.eigenvals(4)

print(eigenval)
print(eigenval[1]-eigenval[0])
```

```python
def anharmonicity(tras):
    eig = tras.eigenvals(3)
    return (eig[2]-eig[1])-(eig[1]-eig[0])
```

```python
anar = anharmonicity(transmon)

print(anar)
```
